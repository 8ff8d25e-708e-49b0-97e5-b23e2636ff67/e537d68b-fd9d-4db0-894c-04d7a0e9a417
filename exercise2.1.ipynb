```vadalog
@bind("isParentOf", "csv useHeaders=true", "./disk/kgs", "parents.csv").
@bind("person", "csv useHeaders=true", "./disk/kgs", "persons.csv").

isAncestorOf(ParentID,ChildrenId) :- isParentOf(ParentID,ChildrenId).
isAncestorOf(ParentID,ChildrenId) :- isAncestorOf(ParentID,X), isParentOf(X,ChildrenId).
isAncestorWithD(PName,PSurname,CName,CSurname) :- isAncestorOf(ParentID,ChildrenId),person(ParentID,PName,PSurname,_),person(ChildrenId,CName,CSurname).
hasAncestors(PersonId,Count) :- isAncestorOf(AncestorId,PersonId), Count=mcount(AncestorId).
hasDescendants(AncestorId,Count) :- isAncestorOf(AncestorId,PersonId), Count=mcount(PersonId).


numDescendantsAncestors(Person,NumD,NumA) :- hasAncestors(Person,NumA),hasDescendants(Person,NumD).
numDescendantsAncestors(Person,0,NumA) :- hasAncestors(Person,NumA).
numDescendantsAncestors(Person,NumD,0) :- hasDescendants(Person,NumD).

@output("isAncestorWithD").
@output("hasAncestors").
@output("numDescendantsAncestors").
@output("hasDescendants").
@post("isAncestorWithD","limit(5)").
@relaxedSafety.
```
