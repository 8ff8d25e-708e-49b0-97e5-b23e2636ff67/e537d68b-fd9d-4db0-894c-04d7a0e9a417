For each registry office, compute how many couples have registered their marriage and divorce in the same Registry Office.

```vadalog
@bind("marriage", "csv", "./disk/kgs", "marriages.csv").
@bind("divorce", "csv", "./disk/kgs", "divorces.csv").

sameOff(Id1,Id2,OfficeId) :- marriage(Id1,Id2,_,OfficeId), divorce(Id1,Id2,_,OfficeId).
officeHasCouple(OfficeId,Count) :- sameOff(Id1,Id2,OfficeId),Count=mcount(1).
@output("officeHasCouple").
@post("officeHasCouple","orderBy(1)").
```

```vadalog
@bind("marriage", "csv", "./disk/kgs", "marriages.csv").
@bind("divorce", "csv", "./disk/kgs", "divorces.csv").
@bind("person", "csv", "./disk/kgs", "persons.csv").
% @model("person", "['first:string', 'second:string', 'third:double', 'fourth:string']").

spouse(Id) :- marriage(Id,_,_,_).
spouse(Id) :- marriage(_,Id,_,_).
divorced(Id) :- divorce(Id,_,_,_).
divorced(Id) :- divorce(_,Id,_,_).

marriedNeverDivorced(Name,Surname) :- person(Id,Name,Surname), spouse(Id), not divorced(Id).
@output("marriedNeverDivorced").
@post("marriedNeverDivorced","orderBy(1,2)").
```

Comput the persons (with all the details e.g. name, surname...) who got married and divorced more than twice

```vadalog
@bind("marriage", "csv", "./disk/kgs", "marriages.csv").
@bind("divorce", "csv", "./disk/kgs", "divorces.csv").
@bind("person", "csv", "./disk/kgs", "persons.csv").

marriedAll(Id1,Id2,Date,Office) :- marriage(Id1,Id2,Date,Office).
marriedAll(Id1,Id2,Date,Office) :- marriage(Id2,Id1,Date,Office).
divorcedAll(Id1,Id2,Date,Office) :- divorce(Id1,Id2,Date,Office).
divorcedAll(Id1,Id2,Date,Office) :- divorce(Id2,Id1,Date,Office).

marriedCount(Id,MC) :- marriedAll(Id,_,Date,Office), MC=mcount(Date).
divorcedCount(Id,MC) :- marriedAll(Id,_,Date,Office), MC=mcount(Date).
personMDTId(Id) :- marriedCount(Id,MC),marriedCount(Id,DC), MC>2, DC>2.
personMDT(N,S) :- person(Id,N,S), personMDTId(Id).
@output("personMDT").
@post("personMDT","orderBy(1,2)").
```
